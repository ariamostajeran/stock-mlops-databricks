```python
signals_df = spark.table("p2_signals").toPandas()
signals_df.head()
```

```python
backtester = PortfolioBacktester(
    transaction_cost_per_trade=0.0005,
    max_positions_per_day=3,
    weighting_method="confidence"
)

portfolio_df = backtester.backtest(signals_df)
portfolio_summary_df = backtester.summary(portfolio_df)
portfolio_by_stock_df = backtester.by_stock(signals_df)

portfolio_summary_df
```

```python
portfolio_by_stock_df
```

```python
spark.createDataFrame(portfolio_df) \
    .write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("p2_portfolio_backtest_daily")

spark.createDataFrame(portfolio_summary_df) \
    .write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("p2_portfolio_backtest_summary")

spark.createDataFrame(portfolio_by_stock_df) \
    .write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("p2_portfolio_backtest_by_stock")
```
